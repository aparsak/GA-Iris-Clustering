```python
def euclidean_distance(X, Y):
    return np.sqrt(np.sum(np.power(X - Y, 2), axis=1))
```

```python
def calculate_fitness(chromosome, distance):
    cluster_distances = []

    for cluster_num in range(1, 4):
        centroid_index = np.random.choice(len(chromosome), size=1)[0]
        cluster_indices = np.where(chromosome == cluster_num)[0]
        distances_to_centroid = abs(distance[centroid_index] -  distance[cluster_indices])
        fitness_cluster = 1 / np.sum(distances_to_centroid)
        cluster_distances.append(fitness_cluster)

    overall_fitness = np.sum(cluster_distances)

    return overall_fitness


'''for i in range(0 , len(population)):
    fitness_scores = [calculate_fitness(population[i], distance)]
    print(fitness_scores)
'''

amounts = []

for i in range(len(population)):
    fitness_score = calculate_fitness(population[i], distance)
    amounts.append(fitness_score)

for i, amount in enumerate(amounts):
    print(f"Population [{i}]: Fitness Score = {amount}")


population_fitness_pairs = {f"Population [{i}]": amount for i, amount in enumerate(amounts)}

```

```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

def crossover(parent_1, parent_2, parent_3):
    new_chromosome = []
    for i in range(len(parent_1)):
        chance = np.random.randint(0, 100)
        if chance <= 50:
            new_chromosome.append(parent_1[i])
        elif chance <= 85:
            new_chromosome.append(parent_2[i])
        else:
            new_chromosome.append(parent_3[i])

    return new_chromosome

def mutate(chromosome, mutation_rate=0.1):
    mutated_chromosome = []
    for gene in chromosome:
        mutation_chance = np.random.rand()
        if mutation_chance < mutation_rate:
            mutated_gene = np.random.randint(0, 10)
            mutated_chromosome.append(mutated_gene)
        else:
            mutated_chromosome.append(gene)
    return mutated_chromosome

def report_clustering(dataframe):
    correct = sum(dataframe['labels'] == dataframe['variety'])

    print('Total accuracy: ', correct / len(dataframe))
    print('------------------------------------')
    print('Count of correct predictions: ', correct)
    print('Count of wrong predictions: ', len(dataframe) - correct)

    class_correct = {class_label: 0 for class_label in dataframe['variety'].unique()}
    class_wrong = {class_label: 0 for class_label in dataframe['variety'].unique()}

    for class_label in class_correct.keys():
        mask = dataframe['variety'] == class_label
        class_correct[class_label] = sum(dataframe.loc[mask, 'labels'] == class_label)
        class_wrong[class_label] = sum(dataframe.loc[mask, 'labels'] != class_label)

    print('------------------------------------')
    for class_label in class_correct.keys():
        print(f'{class_label} correct: ', class_correct[class_label])
        print(f'{class_label} wrong: ', class_wrong[class_label])
        print('------------------------------------')

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['variety'] = iris.target_names[iris.target]

all_indices = list(range(len(df)))

# Example clustering loop
num_iterations = 1000  # Increase the number of iterations
best_chromosome = None
best_accuracy = 0.0

for iteration in range(num_iterations):
    index3 = np.random.choice(all_indices)
    index1, index2, index4, index5 = np.random.choice(all_indices, size=4, replace=False)

    parent_3 = df.iloc[index3].values[:-1]  # Exclude the last column ('variety') for clustering
    parent_1 = df.iloc[index1].values[:-1]
    parent_2 = df.iloc[index2].values[:-1]

    child1 = crossover(parent_1, parent_2, parent_3)
    child2 = crossover(parent_1, parent_2, parent_3)

    # Apply mutation to a random chromosome
    random_chromosome_index = np.random.choice(all_indices)
    random_chromosome = df.iloc[random_chromosome_index].values[:-1]
    mutated_chromosome = mutate(random_chromosome)

    df.iloc[random_chromosome_index, :-1] = mutated_chromosome

    # Assign clustering labels based on the best chromosome
    df['labels'] = best_chromosome if best_chromosome is not None else np.random.choice([1, 2, 3], size=len(df))

    # Calculate accuracy and update best chromosome if needed
    current_accuracy = sum(df['labels'] == iris.target) / len(df)
    if current_accuracy > best_accuracy:
        best_accuracy = current_accuracy
        best_chromosome = df['labels'].copy()

    print(f"\nIteration {iteration + 1}:")
    print("Current Clustering Accuracy:", current_accuracy)


```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

kmeans = KMeans(n_clusters=3, random_state=42)
data['labels'] = kmeans.fit_predict(data_scaled)

def report_kmeans_clustering(dataframe):

    correct = sum(dataframe['labels'] == iris.target)

    print('Total accuracy: ', correct / len(dataframe))
    print('------------------------------------')
    print('Count of correct predictions: ', correct)
    print('Count of wrong predictions: ', len(dataframe) - correct)

    class_correct = {class_label: 0 for class_label in range(3)}
    class_wrong = {class_label: 0 for class_label in range(3)}

    for class_label in class_correct.keys():
        mask = dataframe['labels'] == class_label
        class_correct[class_label] = sum(dataframe.loc[mask, 'labels'] == iris.target[mask])
        class_wrong[class_label] = sum(dataframe.loc[mask, 'labels'] != iris.target[mask])

    print('------------------------------------')
    for class_label in class_correct.keys():
        print(f'Class {class_label} correct: ', class_correct[class_label])
        print(f'Class {class_label} wrong: ', class_wrong[class_label])
        print('------------------------------------')

pca = PCA(n_components=2)
data_pca = pd.DataFrame(pca.fit_transform(data_scaled), columns=['PCA1', 'PCA2'])
data_pca['labels'] = data['labels']

plt.figure(figsize=(8, 6))
for label in data['labels'].unique():
    plt.scatter(data_pca[data_pca['labels'] == label]['PCA1'],
                data_pca[data_pca['labels'] == label]['PCA2'],
                label=f'Cluster {label}', alpha=0.7)

plt.scatter(pca.transform(kmeans.cluster_centers_)[:, 0],
            pca.transform(kmeans.cluster_centers_)[:, 1],
            marker='X', s=200, c='red', label='Centroids')

plt.title('K-Means Clustering on Iris Dataset')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.legend()
plt.show()

# Report clustering performance
print("\nK-Means Clustering Performance:")
report_kmeans_clustering(data)
```
